```python
import pickle
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
nltk.download("wordnet")
word_lema= WordNetLemmatizer()
from nltk.corpus import stopwords
nltk.download("stopwords")
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
model5= pickle.load(open("model.pkl","rb"))
vector = pickle.load(open("vec.pkl","rb"))
```

```python
text= "You've Been Selected! Claim Your $500 Gift Card Now - Limited Time!"
```

```python
stop_word = set(stopwords.words("english"))
```

```python
import re
def normalize(text):
    text= text.lower()
    # multiple letter
    text= re.sub(r'(.)\1+',r'\1', text)
    # punction remove
    text= re.sub(r'[^\w\s]',"", text)
    # for digit remove
    text = re.sub(r"\d", "", text)
    # constractions
    text = re.sub(r"can't","cannot", text)
    text = re.sub(r"isn't","isnot", text)
    # remove bengali word
    #text = re.sub(r'\P{L}*\p{Bengali}+(?:\P{L}+\p{Bengali}+)*\P{L}*', "", text)
    # url remove 
    text= re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+',"",text)
    
    token= word_tokenize(text)
    text = [word_lema.lemmatize(word) for word in token if word not in stop_word]
    text = " ".join(text)
    return text
```

```python
text= normalize(text)
```

```python
import pandas as pd
df1= pd.Series(text)
```

```python
tst= vector.transform(df1)
tst.shape
```
